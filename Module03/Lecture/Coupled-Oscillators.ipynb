# Eigenvalues and Eigenvectors: The Coupled Oscillator 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Two coupled masses

<div>

</div>

From Newton's and Hooke's Laws, the net force on the two masses is:
$$ m_1 a_1 = -k_1 x_1 + k_2(x_2 - x_1),$$
$$ m_2 a_2 = -k_2(x_2 - x_1) -k_3x_2$$

Rearrange equations so that common displacements (e.g. $x_1,\;x_2$) are grouped together:
$$ \;\; m_1 a_1 = -(k_1 + k_2)x_1 + k_2x_2 \;\;(1),$$
$$ m_2 a_2 = k_2x_1 -(k_2 + k_3)x_2 \;\;\;\;(2).$$

The equation of motion for oscillations is:
$$m a = m \frac{d^2x}{dt^2} = -kx,$$
and has a solution
$$x(t) = A\cos(\omega t) + B\sin(\omega t) = C\cos(\omega t + \delta)$$

We can rewrite eqs. $(1)$ and $(2)$ into martix form, with

$$\bf{x}=
\begin{pmatrix}
x_1(t)\\
x_2(t)
\end{pmatrix},\;\;
\bf{K} = 
\begin{pmatrix}
(k_1 + k_2)/m_1 & -k_2/m_1\\
-k_2/m_2        & (k_2 + k_3)/m_2
\end{pmatrix},
$$

Leading to the matrix equation:
$$\frac{d{\bf{x^2}}}{dt^2} = -{\bf{K x}}\;\;,\;\;(3)$$

$$-\omega^2 {\bf{A}} = -{\bf{K A}}\;\;,\;\;(4)$$

where $\bf{A}$ is a matrix containing the amplitude constant:
$$ \bf{A} = 
\begin{pmatrix}
A_1\\
A_2
\end{pmatrix}
$$

Rearranging:

$$({\bf{K}} - \omega^2{\bf{I_2}}){\bf{A}}= 0,$$

where ${\bf{I_2}}$ is the 2D identity matrix:
$${\bf{I_2}} = 
\begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix}
$$

and the characteristic equation is:
$$|{\bf{K}} - \omega^2{\bf{I_2}}|{\bf{A}}= 0,\;\; (5)$$

Eq. 5 is a quadratic for the angular frequency (e.g. $\omega^2$) and so will have two solutions. The two frequencies which the system can sinusoidally oscillate with are called the **normal frequencies** and are given by the **eigenvalues**. The corresponding **eigenvectors** are known as the normal modes. 

### Identical Springs and Equal Masses
For simplisity in the example, we will assume that all three springs are identical, as well as the two masses:
$$k_1 = k_2 = k_3 = k,$$
$$m_1 = m_2 = m$$

The matrix ${\bf{K}}$ then becomes:

$$
{\bf{K}} = 
\begin{pmatrix}
2k/m & -k/m\\
-k/m & 2k/m
\end{pmatrix}
$$

### Eigenvectors:
The characteristic equation is then:

$$
|{\bf{K}}-\omega^2{\bf{I_2}}| = 
\begin{vmatrix}
2k/m - \omega^2 & -k/m\\
-k/m            & 2k/m - \omega^2 
\end{vmatrix} = (2k/m - \omega^2)(2k/m - \omega^2) - (-k/m)(-k/m) = (2k/m - \omega^2)^2 - k^2/m^2 = (k/m-\omega^2)(3k/m - \omega^2) = 0
$$

$$\omega_1 = \sqrt{\frac{3k}{m}},\;\; \omega_2 = \sqrt{\frac{k}{m}}$$  

### Eignevalues:

Subbing the eignevalues into the matrix equation allows us to find the eigenvectors.

${\bf{\omega_1}}:$ First Noraml Mode

$$
({\bf{K}} - \omega^2{\bf{I_2}}) = 
\begin{pmatrix}
-k/m & -k/m\\
-k/m & -k/m
\end{pmatrix}
$$

$$
({\bf{K}} - \omega^2{\bf{I_2}}){\bf{A}} = 0 = 
(-k/m)\begin{pmatrix}
1 & 1\\
1 & 1
\end{pmatrix}
\begin{pmatrix}
A_1\\
A_2
\end{pmatrix}
$$

$$A_1 + A_2 = 0$$
$$\Rightarrow
{\bf{A}} = 
\begin{pmatrix}
1\\
-1
\end{pmatrix}
$$

**Solution:**
$$
{\bf{x}}(t) = 
\begin{pmatrix}
x_1(t)\\
x_2(t)
\end{pmatrix}
=
A_1\begin{pmatrix}
1\\
-1
\end{pmatrix}
\cos(\omega_1 t + \delta),\;\;
$$

$$x_1 (t) =  A_1\cos(\omega_1 t + \delta)$$
$$\;\;x_2 (t) = -A_1\cos(\omega_1 t + \delta)$$

Where $A_1$ and $\delta$ can be obtained from initial conditions of the specific problem.

${\bf{\omega_2}}:$ Second Normal Mode

Following the same procedure as above, one obtains the eigenvector:
$$
\begin{pmatrix}
1\\
1
\end{pmatrix}
$$

**Solution:**
$$
{\bf{x}}(t) = 
\begin{pmatrix}
x_1(t)\\
x_2(t)
\end{pmatrix}
=
A_1\begin{pmatrix}
1\\
1
\end{pmatrix}
\cos(\omega_1 t + \delta),\;\;
$$

$$x_1 (t) =  A_1\cos(\omega_2 t + \delta)$$
$$x_2 (t) =  A_1\cos(\omega_2 t + \delta)$$



### Solving for the eigenvalues and eigenvectors in Python

Assume 
$$ A_1 = 1\;m \\
   \delta = 0\; \text{rad}\\
   k = 1\; \frac{N}{m}\\
   m = 1\; kg\\
   t_0 = 0\;s\\
   t_f = 25\;s
$$

```python
t0 = 0
tf = 50
n = 500

k = 1 # N/m
k1 = k
k2 = k
k3 = k

m = 1 #kg
m1 = m
m2 = m
m3 = m
```

```python
K = np.array([[(k1+k2)/m1,-k2/m1],[-k2/m2,(k2+k3)/m2]])
print("\nK\n",K)
```

### Mode 1: $\omega^2 = 3k/m$

```python
w,v = np.linalg.eig(K)
print('E-value:', w)
print('E-vector', v)
```

```python
t = np.linspace(t0,tf,n)
x1_1 = v[0,0]*np.cos(np.sqrt(w[0])*t)
x2_1 = v[1,0]*np.cos(np.sqrt(w[0])*t)


fig = plt.figure('$\omega^2 = 3$')
ax = fig.add_axes([0.1,0.1,0.8,0.8])

#draw x and v
ax.plot(t,x1_1,'b', label=r'$x_1(t)$')
ax.plot(t,x2_1,'r', label=r'$x_2(t)$')

ax.legend()
ax.set_xlabel('t (s)')
ax.grid();
```

For the case with 3 identical springs and two identical masses, the two masses oscillate back and forth with equal amplitudes, but exactly out of phase, so that $x_1(t) = -x_2(t)$ at all times

### Mode 2: $\omega^2 = k/m$

```python
x1_2 = v[0,1]*np.cos(np.sqrt(w[1])*t)
x2_2 = v[1,1]*np.cos(np.sqrt(w[1])*t)


fig = plt.figure('$\omega^2 = 1$')
ax = fig.add_axes([0.1,0.1,0.8,0.8])

#draw x and v
ax.plot(t,x1_2,'b', label=r'$x_1(t)$')
ax.plot(t,x2_2,'r--', label=r'$x_2(t)$')

ax.legend()
ax.set_xlabel('t (s)')
ax.grid();
```

For the case with 3 identical springs and two identical masses, the two masses oscillate back and forth with equal amplitudes and are exactly in phase, so that $x_1(t) = x_2(t)$ at all times. This means that the middle spring is neirhter stretched or compressed.

## General Solution

A mixture of both modes are needed to explain a general oscillation of the two mass, three spring system. In general one can have:

$$
{\bf{x}}(t) = 
B\begin{pmatrix}
1\\
-1
\end{pmatrix}
\cos(\omega_1 t + \delta_1)
+ C
\begin{pmatrix}
1\\
1
\end{pmatrix}
\cos(\omega_2 t + \delta_2)
$$

$$
= B_1\begin{pmatrix}
1\\
-1
\end{pmatrix}
\cos(\omega_1 t)
+ B_2\begin{pmatrix}
1\\
-1
\end{pmatrix}
\sin(\omega_1 t)
+ C_1\begin{pmatrix}
1\\
1
\end{pmatrix}
\cos(\omega_2 t)
+ C_2\begin{pmatrix}
1\\
1
\end{pmatrix}
\sin(\omega_2 t)
$$

To determine the constants: $B_1, B_2, C_1, C2$ we need to use initial conditions. 
Given the **initial conditions**:
$${\bf{x}}(t = 0) = 
\begin{pmatrix}
1\\
0
\end{pmatrix},\;\;
{\bf{v}}(t=0) = 
\begin{pmatrix}
0\\
0
\end{pmatrix}
$$


$$
{\bf{x}}(0) = 
B_1\begin{pmatrix}
1\\
-1
\end{pmatrix}
+ C_1
\begin{pmatrix}
1\\
1
\end{pmatrix}
=
\begin{pmatrix}
1\\
0
\end{pmatrix}
$$

$$\rightarrow B_1 + C_1 = 1$$ 
$$\rightarrow -B_1 + C_1 = 0$$
$$\Rightarrow B_1 = C_1 = \frac{1}{2}$$

$$
{\bf{v}}(0) = 
\omega_1 B_2\begin{pmatrix}
1\\
-1
\end{pmatrix}
+ \omega_2 C_2
\begin{pmatrix}
1\\
1
\end{pmatrix}
=
\begin{pmatrix}
0\\
0
\end{pmatrix}
$$

$$\rightarrow \omega_1 B_2 + \omega_2 C_2 = 0$$ 
$$\rightarrow -\omega_1 B_2 + \omega_2 C_2 = 0$$
$$\Rightarrow B_2 = C_2 = 0$$

```python
x1_gen = 0.5*x1_1 + 0.5*x1_2
x2_gen = 0.5*x2_1 + 0.5*x2_2
```

```python
fig = plt.figure('General')
ax = fig.add_axes([0.1,0.1,0.8,0.8])

#draw x and v
ax.plot(t,x1_gen,'b', label=r'$x_1(t)$')
ax.plot(t,x2_gen,'r', label=r'$x_2(t)$')

ax.legend()
ax.set_xlabel('t (s)')
ax.grid();
```
